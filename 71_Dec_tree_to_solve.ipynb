## Decision Tree for Regression

```python
path='https://frenzy86.s3.eu-west-2.amazonaws.com/fav/tecno/petrol_consumption.csv'
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
df = pd.read_csv(path)
df
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True)
```

### Regression

```python
X = df.drop(columns='Petrol_Consumption')
y = df['Petrol_Consumption']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 667
                                                    )
```

```python
model = DecisionTreeRegressor()
model.get_params()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test).astype(int)
df = pd.DataFrame({'reale': y_test, 'predetto': y_pred})
df
```

```python
length = y_pred.shape[0] #  
x = np.linspace(0,length,length)

plt.figure(figsize=(10,7))
plt.plot(x, y_test, label='real y')
plt.plot(x, y_pred, label="predicted y'")
plt.legend(loc=2);
```

### Evaluating the Algorithm
To evaluate performance of the regression algorithm, the commonly used metrics are mean absolute error, mean squared error, and root mean squared error. 

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y_true=y_test, y_pred=y_pred)
mse = mean_squared_error(y_true=y_test, y_pred=y_pred)
rmse = np.sqrt(mse)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
```
