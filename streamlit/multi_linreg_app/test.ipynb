```python
import streamlit as st
import pandas as pd
import plotly.express as px
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("https://frenzy86.s3.eu-west-2.amazonaws.com/python/data/Startup.csv")
```

```python
df
```

```python
X = df.drop(columns='Profit')
y = df['Profit']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.2, 
                                                    random_state = 667
                                                    )
```

```python
model = LinearRegression(fit_intercept=True)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
l = y_test_pred.shape[0]
x = np.linspace(0,l,l)


plt.plot(x, y_test_pred, label='y predicted')
plt.plot(x, y_test, label='y target')
plt.title('y prediction - y target in test set')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2score = r2_score(y_test, y_test_pred)
mae = mean_absolute_error(y_test, y_test_pred)
mse = mean_squared_error(y_test, y_test_pred)
rmse = mean_squared_error(y_test, y_test_pred, squared=False)
```
