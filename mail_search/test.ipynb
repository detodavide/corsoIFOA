```python

import pandas as pd
import regex as re
```

```python
df = pd.read_excel("https://frenzy86.s3.eu-west-2.amazonaws.com/IFAO/email_re.xlsx")
```

```python
df
```

```python
list1 = df['content'].to_list()
```

```python
list1 = [str(i) for i in list1]
```

```python
string_of_text = ' '.join(list1).lower()
```

```python
# pattern = r'[\w]+@[\w]+\.[\w]{2,8}'
pattern = r'[a-z0-9]+@[a-z0-9]+\.[a-z0-9]{2,8}'
match = re.findall(pattern, string_of_text)
```

```python
match_series = pd.Series(match)
```

```python
match_series = match_series.unique()
```

```python
df_match = pd.DataFrame(match_series)
```

```python
df_match.columns = ['patterns']
```

```python
df_match.to_csv("match_list.csv", index=False)
```

```python
patterns = df_match['patterns'].to_list()
```

```python
patterns = [str(i) for i in patterns]
```

```python
df_counts = pd.DataFrame()
```

```python
for pattern in patterns:
    df_counts[pattern] = df['content'].astype(str).apply(lambda x: len(re.findall(pattern, x)))

```

```python
dict_match = {}
for col in df_counts.columns:
    
    dict_match[col] = df_counts[col].sum()
```

```python
dict_match = {}
for pattern in patterns:

    match = re.findall(pattern, string_of_text.lower())
    dict_match[pattern] = len(match)
        
```

```python
df_match_count = pd.DataFrame(dict_match, index=[0])
```

```python
df_match_count = df_match_count.T
df_match_count.columns = ['Count']
```

```python
df_match_count = df_match_count.sort_values('Count', ascending=False)
```
